```python
sns.set(style = "white" , color_codes = True)
```

```python
iris = datasets.load_iris()
```

```python
df = pd.read_csv("iris.csv")
```

```python
df
```

```python
df.head()
```

```python
df.plot(kind = "scatter" ,x="sepal.length", y = "sepal.width" )
```

```python
sns.jointplot(x="sepal.length", y = "sepal.width", data = df , size = 5)
```

```python
sns.FacetGrid(df, hue = "variety",size = 5).map(plt.scatter, "sepal.length" , "sepal.width").add_legend()
```

```python
sns.boxplot(x="petal.length", y = "petal.width", data = df[:200])
```

```python
ax = sns.boxplot(x="sepal.length", y = "sepal.width", data = df[:200])
```

```python
ax = sns.stripplot(x="variety", y="petal.length", data = df, jitter = True , edgecolor="gray")
```

```python
sns.violinplot(x="variety", y="petal.length", data = df, size = 6)
```

```python
sns.FacetGrid(df, hue = "variety",size = 6).map(sns.kdeplot, "petal.length").add_legend()
```

```python
sns.pairplot(df, hue = "variety", size = 3,diag_kind = "kde")
```

```python
df.boxplot(by="variety", figsize = (12,6))
```
